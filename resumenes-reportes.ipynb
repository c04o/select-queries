1. Selecciono datos como el ID del cliente; y nombre de la compania como Cliente, nombre del producto, su categoria, nombre de la compan~i
2. Calculo la suma de las ordenes, UnidadesTotales, TotalesPedidos
3. Calculo promedios de pedidos (maximo y minimo)
4. Calculo valores totales
5. Uno Clientes con Ordenes para solo ver los Clientes que tengan Ordenes de las tablas
6. Uno las Ordenes con los Productos que coincidan en ID en las tablas
7. Uno Categorias con las que tengan ID en las tablas
8. Unirlas y ordenarlas

```sql
SELECT 
    c.CustomerID,
    c.CompanyName AS Cliente,
    p.ProductID,
    p.ProductName AS Producto,
    cat.CategoryName AS Categoria,
    sup.CompanyName AS Proveedor,
    SUM(od.Quantity) AS UnidadesTotales,
    COUNT(DISTINCT o.OrderID) AS TotalPedidos,
    AVG(od.Quantity) AS PromedioPorPedido,
    MAX(od.Quantity) AS MaximoPorPedido,
    MIN(od.Quantity) AS MinimoPorPedido,
    SUM(od.Quantity * od.UnitPrice) AS ValorTotal
FROM 
    Customers c
INNER JOIN 
    Orders o ON c.CustomerID = o.CustomerID
INNER JOIN 
    [Order Details] od ON o.OrderID = od.OrderID
INNER JOIN 
    Products p ON od.ProductID = p.ProductID
INNER JOIN 
    Categories cat ON p.CategoryID = cat.CategoryID
INNER JOIN 
    Suppliers sup ON p.SupplierID = sup.SupplierID
GROUP BY 
    c.CustomerID, 
    c.CompanyName,
    p.ProductID, 
    p.ProductName,
    cat.CategoryName,
    sup.CompanyName
ORDER BY 
    c.CompanyName ASC,
    UnidadesTotales DESC;
```
